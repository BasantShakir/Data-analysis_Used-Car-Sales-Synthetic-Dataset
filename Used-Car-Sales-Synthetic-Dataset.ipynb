```python
import matplotlib.pyplot as plt
import seaborn as sns

correlation_matrix = df.corr(numeric_only=True)

print("Correlation Matrix:")
display(correlation_matrix)

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix,
            annot=True,
            cmap='Blues',
            fmt=".2f",
            linewidths=0.5,
            cbar=True)

plt.title('Correlation Matrix for Car Sales Features', fontsize=14, fontweight='bold', pad=15)
plt.tight_layout()
plt.show()
```

```python
plt.scatter(df['Mileage-KM'], df['Sold Price-$'], color='royalblue', alpha=0.5)

plt.title('Station 1: How Mileage Kills the Car Value')
plt.xlabel('Mileage KM')
plt.ylabel('Sold Price')
plt.show()
```

```python
#Fuel Type vs Margin
import matplotlib.pyplot as plt

energy_margin = df.groupby('Energy')['Margin-%'].mean()
energy_margin.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Station 2: Which Energy Type Makes More Profit?')
plt.xlabel('Energy Type')
plt.ylabel('Average Margin %')
plt.show()
```

```python
# Station 3: Sales Rating vs Commission
import matplotlib.pyplot as plt

rating_comm = df.groupby('Sales Rating')['Sales Commission-$'].mean()
rating_comm.plot(kind='bar', color='darkblue', edgecolor='black')

plt.title('Station 3: Customer Rating vs Sales Commission')
plt.xlabel('Sales Rating (1-5)')
plt.ylabel('Average Commission $')
plt.show()
```

```python
# Block 2: Simple Bar Plot for Ratings
import matplotlib.pyplot as plt

# بنجمع الداتا حسب التقييم ونجيب متوسط العمولات
rating_data = df.groupby('Sales Rating')['Sales Commission-$'].sum()

# بنرسم الأعمدة بلون أزرق
rating_data.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Commissions by Sales Rating')
plt.xlabel('Sales Rating')
plt.ylabel('Sum of Commission')

plt.show()
```

```python
# Block 3: Simple Bar Plot for Energy
import matplotlib.pyplot as plt

# بنجيب متوسط نسبة الربح لكل نوع طاقة
energy_data = df.groupby('Energy')['Margin-%'].mean()

# بنرسم الأعمدة
energy_data.plot(kind='bar', color='darkblue', edgecolor='black')

plt.title('Profit Margin by Energy Type')
plt.xlabel('Energy Type')
plt.ylabel('Average Margin %')

plt.show()
```
